```python
import pandas as pd
import requests
```

```python
url = "https://tradingview18.p.rapidapi.com/symbols/auto-complete"
```

```python
querystring = {"query": "tela"}
```

```python
headers = {
    "x-rapidapi-host": "tradingview18.p.rapidapi.com",
    "x-rapidapi-key": "73c3b12d1bmsh7de9b3b75d8a156p1225e3jsn81e4359ac1e7"
}
```

```python
response = requests.get(url, headers=headers, params=querystring)
```

```python
data = response.json()
print(data)
```

```python
quotes_df = pd.DataFrame(data['quotes'])[
    ['symbol','shortname','longname','exchange','exchDisp','sector','industry']
]

print(quotes_df.head())
```

```python
news_df = pd.DataFrame(data['news'])[
    ['uuid','title','publisher','link','providerPublishTime','relatedTickers']
]
news_df
```

```python
quotes_df.to_csv("tradingview_quotes.csv", index=False)
news_df.to_csv("tradingview_news.csv", index=False)
```
